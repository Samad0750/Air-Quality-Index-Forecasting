## It Shows 0 non null values but according to dataset missing values are tagged with -200. This is described in UCI repo. documentation.

```python
## counting the number of times -200 qppears in the table.
df.isin([-200]).sum(axis=0)
```

```python
df.replace(-200,np.NaN,inplace=True)
```

```python
column_means = df.iloc[:, 2:].mean() 
# Fill null values with respective column means
df.iloc[:, 2:] = df.iloc[:, 2:].fillna(column_means)
```

## For Forecasting we use 'FB PROPHET' library.


```python
!pip install prophet
from prophet import Prophet
```

```python
model=Prophet()
```

```python
model.fit(df_copy)
```

```python
future=model.make_future_dataframe(periods=365,freq='H')
```

```python
df_copy.head()
```

```python
future.head()
```

```python
future.tail()
```

```python
forecast=model.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
forecast[['ds','yhat','yhat_lower','yhat_upper']].head()
```

```python
fig1=model.plot(forecast)
```

```python
fig2=model.plot_components(forecast)
```
